### REGRESIÓN LINEAL

```python
import matplotlib.pyplot as plt
#función que nos permite hace una regresión lineal simple
def rls(x:list[float],y:list[float]) -> list[float]:

    suma_x = sum(x)
    suma_y = sum(y)

    xy = [xi * yi for xi, yi in zip(x,y)]
    suma_xy = sum(xy)

    x2 = [xi ** 2 for xi in x]
    suma_x2 = sum(x2)

    alfa = ((n*suma_xy)-(suma_x * suma_y)) / ((n * suma_x2) - (suma_x**2))
    beta = suma_y-(alfa*suma_x)
    #beta = (suma_y / n) - alfa * (suma_x / n)
    
    return[alfa, beta]
```

```python
#Clase 12 febrero 2026
yi= [23,37, 45, 46, 58, 67, 79, 80, 89] #chirridos de grillo
xi= [64, 50, 52, 51, 57, 55, 60, 63, 63] #temperatura 
n= 9
print (rls(xi,yi))

#Cheatsheet para hacer ploteo de la regresión lineal simple
alfa, beta = rls(xi, yi)
print("Pendiente:", alfa)
print("Intercepto:", beta)

#ecuacion de regresión lineal y= a + bx
y_pred = [alfa * x + beta for x in xi]

# gráfica
plt.scatter(xi, yi)
plt.plot(xi, y_pred)
plt.xlabel("Temperatura")
plt.ylabel("Chirridos")
plt.title("Regresión Lineal Simple")
plt.show()
```
