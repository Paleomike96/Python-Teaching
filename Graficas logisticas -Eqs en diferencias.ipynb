# **Tendencia a un Límite**

**Ejercicio de los Peces**

En la siguiente gráfica se aprecia que el punto de equilibrio 13,333 es el punto de balance donde los peces que nacen son los mismos que los que se pesacan. Al tener menos que el punto de equilibrio, en este caso 10,000, los peces se extinguen en un periodo de 9.92 años, pero, si están por encima de este punto, como 15,000, la población de peces aumenta.

```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros del modelo
a = 1.15
b = -2000
pasos = 21
t = np.arange(pasos)

#Codigo de la funcion
def calcular_poblacion(p0, a, b, n_pasos):
    p = [p0]
    for _ in range(n_pasos - 1):
        proximo = a * p[-1] + b #ecuacion correlacion lineal
        p.append(max(0, proximo)) # Sin poblaciones negativas
    return p


p_10k = calcular_poblacion(10000, a, b, pasos)
p_13k = calcular_poblacion(13333.33, a, b, pasos)
p_15k = calcular_poblacion(15000, a, b, pasos)

plt.figure(figsize=(12, 7), facecolor='#fdfdfd')


plt.plot(t, p_10k, color='aqua', marker='o', linewidth=2, label='P0 = 10,000 (Extinción)')
plt.plot(t, p_13k, color='hotpink', linestyle='--', linewidth=3, label='P0 = 13,333 (Límite)')
plt.plot(t, p_15k, color='lime', marker='s', linewidth=2, label='P0 = 15,000 (Crecimiento)')

# Punto de extinción calculado (9.92)
plt.annotate('Extinción ≈ 9.92 años', xy=(9.92, 0), xytext=(12, 5000),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1))


plt.title('Dinámica de Población de Peces', fontsize=16, fontweight='bold')
plt.xlabel('Años (n)', fontsize=12)
plt.ylabel('Población (Pn)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.axhline(0, color='black', linewidth=1.5)
plt.legend(loc='upper left', frameon=True, shadow=True)

plt.show()
```

**Ejercicio ecuación logística**

En esta sección aplicamos un modelo de crecimiento logístico para determinar el crecimiento poblacional suponiendo 3 casos iniciales: *P_0= 1000*, *P_0= 2500* y *P_0= 5000* siendo menor, igual y mayor a k respectivamente, siendo el valor de *k=2500*, y el coeficiente *r=0.001*

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos nuestros parámetros del modelo de crecimiento logístico
r = 0.001
K = 2500
pasos = 30
t = np.arange(pasos)

# Va nuestra función del modelo logístico
def modelo_logistico(p0, r, K, n_pasos):
    p = [p0]
    for _ in range(n_pasos - 1):
        crecimiento = r * p[-1] * (1 - p[-1]/K)
        proximo = p[-1] + crecimiento
        p.append(proximo)
    return p

# Metemos nuestros tres escenarios
p_menor = modelo_logistico(1000, r, K, pasos)
p_igual = modelo_logistico(2500, r, K, pasos)
p_mayor = modelo_logistico(5000, r, K, pasos)

# Parámetros de nuestra gráfica
plt.figure(figsize=(10,6))
plt.plot(t, p_menor, marker='o', label='P0 = 1000 (< K)')
plt.plot(t, p_igual, linestyle='--', label='P0 = 2500 (= K)')
plt.plot(t, p_mayor, marker='s', label='P0 = 5000 (> K)')

# Línea de capacidad de carga
plt.axhline(K, linestyle=':', label='Capacidad de carga K=2500')
plt.title('Modelo Logístico de Crecimiento Poblacional')
plt.xlabel('Tiempo (n)')
plt.ylabel('Población P(n)')
plt.grid(True)
plt.legend()
plt.show()
```

Van las tres gráficas separadas

```python
#Primera gráfica, P0 menor a k
# Parámetros
plt.figure(figsize=(10,6))
plt.plot(t, p_menor, marker='o', label='P0 = 1000 (< K)')

# Línea de capacidad de carga
plt.axhline(K, linestyle=':', label='Capacidad de carga K=2500')
plt.title('Modelo Logístico de Crecimiento Poblacional')
plt.xlabel('Tiempo (n)')
plt.ylabel('Población P(n)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Segunda gráfica
# Parámetros
plt.figure(figsize=(10,6))
plt.plot(t, p_igual, linestyle='--', label='P0 = 2500 (= K)')

# Línea de capacidad de carga
plt.axhline(K, linestyle=':', label='Capacidad de carga K=2500')
plt.title('Modelo Logístico de Crecimiento Poblacional')
plt.xlabel('Tiempo (n)')
plt.ylabel('Población P(n)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Tercera gráfica, con P0 mayor a k
# Parámetros
plt.figure(figsize=(10,6))
plt.plot(t, p_mayor, marker='s', label='P0 = 5000 (> K)')

# Línea de capacidad de carga
plt.axhline(K, linestyle=':', label='Capacidad de carga K=2500')
plt.title('Modelo Logístico de Crecimiento Poblacional')
plt.xlabel('Tiempo (n)')
plt.ylabel('Población P(n)')
plt.grid(True)
plt.legend()
plt.show()
```

**Problema de Temperatura**

```python
# Datos del problema
T0 = 180       # Temperatura inicial
TA = 80        # Temperatura ambiente
K = -0.05      # Constante calculada (a = 0.95)
pasos = 61     # Graficaremos una hora (60 min) para ver el equilibrio
n = np.arange(pasos)


def enfriamiento_discreto(T0, TA, K, n_pasos):
    t_vals = [T0]
    for _ in range(n_pasos - 1):
        # Ecuación recurrente: Tn+1 = Tn + K(Tn - TA)
        proximo = t_vals[-1] + K * (t_vals[-1] - TA)
        t_vals.append(proximo)
    return t_vals

temperaturas = enfriamiento_discreto(T0, TA, K, pasos)


plt.figure(figsize=(12, 7))


plt.plot(n, temperaturas, color='aqua', marker='o', markersize=4,
         linewidth=1, label='Temperatura del objeto (T_n)')


plt.axhline(TA, color='hotpink', linestyle='--', linewidth=2,
            label=f'Temperatura Ambiente (L = {TA}°)')


plt.annotate(f'Inicio: {T0}°', xy=(0, T0), xytext=(5, T0+5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'10 min: {temperaturas[10]:.2f}°', xy=(10, temperaturas[10]),
             xytext=(15, temperaturas[10]+20),
             arrowprops=dict(facecolor='black', arrowstyle='->'))


plt.title('Modelo  ($T_{n+1} = T_n - 0.05(T_n - 80)$)', fontsize=14)
plt.xlabel('Tiempo (minutos)', fontsize=12)
plt.ylabel('Temperatura (°C)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend()

plt.show()
```

En la gráfica se aprecia que tras 1 hora, el objeto se acerca a la temperatura ambiente, pero tras 120 min, la temperatura llega a 80.20° grados. Al tratarse de una asíntota, nunca alcanzará la temperatura ambiente.
