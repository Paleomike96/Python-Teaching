```python
import numpy as np
import matplotlib.pyplot as plt

#Determinamos nuestros parámetros de acuerdo a los ejercios en clase
k = 1/36
dt = 6
n_max = 20

#Armamos nuestros arreglos de temperaturas
T = np.zeros(n_max)
P = np.zeros(n_max)

#Condiciones iniciales de Temperatura en dos habitaciones
T[0] = 9
P[0] = 33

#Iteramos nuestro sistema en ambas habitaciones de acuerdo a la ecuación
for n in range(n_max - 1):
    T[n+1] = T[n] + k * dt * (P[n] - T[n])
    P[n+1] = P[n] + k * dt * (T[n] - P[n])

#Graficamos los resultados
plt.figure()
plt.plot(T, label='Sala 1: T')
plt.plot(P, label='Sala 2: P')
plt.axhline(21, linestyle='--', label='Equilibrio')
plt.xlabel('n (pasos de tiempo)')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid()
plt.show()
```
