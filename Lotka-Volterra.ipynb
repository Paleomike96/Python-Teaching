Programa el sistema de ecuaciones de Lotka-Volterra, y gráficar $n$ vs $Z(n)$, $n$ vs $C(n)$ y por último una gráfica de $C(n)$ vs $Z(n)$ para los siguentes parámetros.

- **r** = 0.08
- **c** = 0.1
- **b** = 0.01
- **d** = 0.15
- **n** = 100 (100 iteraciones)
- $C_{0}$ = 0.2
- $Z_{0}$ = 0.9


**Sugerencia:** Crea una función que reciba como parámetros **function(r, c, b, d, n, $C_{0}$, $Z_{0}$)**.

1. ¿Qué es lo que interpretas del modelo?
2. Cambia las condiciones iniciales $C_{0}$ y $Z_{0}$ ¿Qué observas?

```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función que modela el sistema de Lotka-Volterra
def lotka_volterra(r, c, b, d, n, C0, Z0):
    # Inicializamos los arrays para las poblaciones de conejos (C) y zorros (Z)
    C = np.zeros(n)  # Array para la población de conejos
    Z = np.zeros(n)  # Array para la población de zorros

    # Asignamos las condiciones iniciales
    C[0] = C0  # Población inicial de conejos
    Z[0] = Z0  # Población inicial de zorros

    # Iteramos para calcular las poblaciones en cada paso
    for i in range(1, n):
        # Ecuaciones de Lotka-Volterra
        C[i] = C[i-1] + (r * C[i-1] - c * C[i-1] * Z[i-1])  # Cambio en conejos
        Z[i] = Z[i-1] + (b * C[i-1] * Z[i-1] - d * Z[i-1])  # Cambio en zorros

    return C, Z

# Parámetros del modelo
r = 0.08  # Tasa de crecimiento de los conejos
c = 0.1   # Tasa de depredación de los zorros
b = 0.01  # Tasa de crecimiento de zorros a partir de conejos
d = 0.15  # Tasa de mortalidad de los zorros
n = 100   # Número de iteraciones
C0 = 0.2  # Población inicial de conejos
Z0 = 0.9  # Población inicial de zorros

# Llamamos a la función para obtener las poblaciones
C, Z = lotka_volterra(r, c, b, d, n, C0, Z0)

# Gráficas
plt.figure(figsize=(15, 5))

# Gráfica de n vs C(n)
plt.subplot(1, 3, 1)
plt.plot(range(n), C, label='Conejos (C)', color='blue')
plt.title('Población de Conejos (C) vs Tiempo')
plt.xlabel('Tiempo (n)')
plt.ylabel('Población de Conejos (C)')
plt.grid()
plt.legend()

# Gráfica de n vs Z(n)
plt.subplot(1, 3, 2)
plt.plot(range(n), Z, label='Zorros (Z)', color='red')
plt.title('Población de Zorros (Z) vs Tiempo')
plt.xlabel('Tiempo (n)')
plt.ylabel('Población de Zorros (Z)')
plt.grid()
plt.legend()

# Gráfica de C(n) vs Z(n)
plt.subplot(1, 3, 3)
plt.plot(C, Z, label='Zorros vs Conejos', color='green')
plt.title('Población de Zorros (Z) vs Conejos (C)')
plt.xlabel('Población de Conejos (C)')
plt.ylabel('Población de Zorros (Z)')
plt.grid()
plt.legend()

# Mostrar las gráficas
plt.tight_layout()
plt.show()
```

### Interpretación del Modelo
Con este modelo, los parámetros de Lotka-Volterra muestran la dinámica entre las poblaciones de conejos y zorros. Con un crecimiento moderado de conejos y una tasa de depredación considerable, se observa que la población de conejos  aumenta, pero a medida que la población de zorros también crece, la depredación reduce rápidamente el número de conejos. Esto lleva a un ciclo en el que las poblaciones oscilan entre sí, mostrando la interrelación entre ambas especies.

### Cambios en las Condiciones Iniciales
Al modificar las condiciones iniciales $ C_0 $ y $ Z_0 $, se pueden observar diferentes dinámicas en el sistema. Si se aumenta $ C_0 $, la población de conejos puede crecer rápidamente, lo que puede llevar a un aumento en la población de zorros a medida que hay más presas disponibles. En contraste, al incrementar $ Z_0 $, la población de conejos disminuye rápidamente debido a la depredación, lo que también afecta la población de zorros a largo plazo.
