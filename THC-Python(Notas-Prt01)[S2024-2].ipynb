## Operadores aritméticos

Los operadores aritméticos disponibles se muestran a continuación:

| Operador 	|        Operación       	|   Uso  	|
|:--------:	|:----------------------:	|:------:	|
|     +    	|          Suma          	|  a + b 	|
|     -    	|          Resta         	|  a - b 	|

La jerarquía de las operadores, es decir, el orden en que se evalúan las operaciones es la siguiente:

* ***Primero***. *Paréntesis*. Aunque no es propiamente un operador, permite cambiar el orden en que se evalúan las operaciones. Para paréntesis anidados, el que está más al interior es el que se evalúa primero.
* ***Segundo***. *Potencias*.
* ***Tercero***. *Multiplicación y división*. (El módulo y la división entera se encuentran también en este nivel)
* ***Cuarto***. *Suma y resta*.

````python
# Suma
3 + 4
````

````python
# Resta
4 - 5
````

````python
# Multiplicacion
3*7
````

````python
# Division
5/6
````

````python
# Potencia
3.3**2
````

````python
# Modulo
5%2
````

````python
# Division entera
11//3
````

En el siguiente ejemplo se muestra como cambia el orden de evaluación de las operaciones al hacer uso de los paréntesis.

````python
# Jerarquia
print((3+4) * 2)
print(3 + 4*2)
````

Cuando se evalua la expresión `(3+4) * 2`, dado que se incluyen los paréntesis, se realiza primero la suma y el resultado (7) se multiplica por 2, dando 14. En la segunda expresión `3 + 4*2`, dado que la multiplicación tiene mayor jerarquía que la suma, se realiza primero la múltiplicación y al resultado (8) se le suma 3, dando como respuesta final 11.


Algo importante a tener en cuenta es que al realizar operaciones con datos numéricos de un mismo tipo se puede producir un tipo de dato numérico diferente. Un ejemplo de esto sucede cuando se realiza la división entre números enteros, ya que el resultado será de tipo `float`.

Para otros tipos de datos también se pueden usar los operadores aritméticos pero en un sentido diferente, por ejemplo, en el caso de la suma de cadenas esta produce concatenación de cadenas.

También se pueden hacer operaciones entre diferentes tipos de datos, pero en algunas ocasiones aparecen errores `TypeError`, que ocurren cuando la operación no está definida entre los tipos de datos involucrados. # Variables

Las variables en Python no necesitan una declaración explícita para reservar espacio en la memoria. La declaración ocurre automáticamente cuando se asigna un valor a una variable, ya sea a través de una literal o mediante otra variable previamente declarada. Para declarar una variable se utiliza el operador de *asignación* `=`. La sintaxis a usar es:

```python
nombre_variable = valor_variable
```

esto es, el operando a la izquierda del operador `=` es el **nombre de la variable** y el operando a la derecha del operador `=` es el **valor almacenado** en la variable.

Dado que Python es un lenguaje interpretado, la asignación de variables es dinámica, por lo que el tipo de dato asignado a una variable se puede cambiar en cualquier momento. Esto requiere poner atención en el manejo de las variables para evitar errores de tipo `TypeError`.

**Nota**: Considere lo siguiente para los nombres de variables:
* El nombre de las variables debe comenzar con una letra, de la `A` a la `Z` o de la `a` a la `z` o un guión bajo (`_`) seguido de cero o más letras, guiones bajos y dígitos (`0` a `9`).
* Python no permite caracteres de puntuación o caracteres como `@`, `$` y `%` dentro de los nombres.
* También, al ser un lenguaje de programación sensible a mayúsculas y minúsculas se debe prestar atención en los nombres ya que, por ejemplo, `Variable` y `variable`, hacen referencia a dos identificadores de variables diferentes.
* Las palabras reservadas (*keywords*) que no se pueden usar como nombres de variables se puede consultar mediante la siguiente instrucción:

````python
help('keywords')
````

## Eliminación de variables

Para eliminar la referencia a una o varias variables se utiliza el comando `del`. La sintaxis a usar es:

```python
del var1[, var2[, var3[..., varN]]]
```

**Nota** La notación con corchetes ([ ]) indica que los valores dentro son argumentos opcionales. Por ejemplo, si tenemos tres variables `var1`, `var2`, `var3`, como en la notación de arriba, usos válidos del comando `del` son:

```python
del var2 # Se borra la referencia a var2
del var3, var1 # Se borra la referencia a var3 y var1
```

````python
# Ejemplo de eliminacion de variables
print(a, b, c, d, e, f)
del a  # Borrar una sola variable
del b, c, d, e, f  # Borrar multiples variables
````

**Nota**: Una vez eliminada una variable ya no será posible acceder a ella a través de su nombre (identificador). Si se quiere hacer referencia a una variable que no este declarada o la cual haya sido eliminada, se obtendrá un error de tipo `NameError`.

## Variables, operadores booleanos (lógicos) y relacionales

Las *variables booleanas* solo pueden tomar los valores `True` o `False`. Los *operadores booleanos o lógicos* (`and`, `or` y `not`) usan operandos de tipo `bool` y regresan valores de este tipo. Los *operadores relacionales* (`==`, `!=`, `>`, `<`, `>=`,`<=`) usan operandos de diferentes tipos, pero siempre regresan un valor de tipo `bool`. La siguiente tabla muestra los operadores booleanos y relacionales junto con su significado:

| Operador 	| Nombre            	|
|:--------:	|-------------------	|
|    and   	| y lógico          	|
|    or    	| o lógico          	|
|    not   	| Negación          	|
|    ==    	| Igual             	|
|    !=    	| Diferentes        	|
|     >    	| Mayor que         	|
|     <    	| Menor que         	|
|    >=    	| Mayor o igual que 	|
|    <=    	| Menor o igual que 	|

Primero se mostrará el uso de los operadores lógicos. En el ejemplo mostrado a continuación, se declararán dos variables booleanas `cero` con el valor `False` y `uno` con el valor `True` para construir las tablas de verdad para los operadores *binarios* `and` y `or`, y para el operador *unario* `not`, con el objetivo de mostrar todos los posibles resultados de aplicar los operadores a las variables booleanas.

````python
# Declaracion de variables booleanas
cero, uno = False, True
print(cero)  # Falso
print(uno)  # Verdadero
````

````python
# Tabla de verdad para el operador binario and
print(cero and cero)
print(cero and uno)
print(uno and cero)
print(uno and uno)
````

````python
# Tabla de verdad para el operador binario or
print(cero or cero)
print(cero or uno)
print(uno or cero)
print(uno or uno)
````

````python
# Tabla de verdad para el operador unario not
print(not cero)
print(not uno)
````

Cuando se usan literales o variables de tipo numérico, es posible utilizar los operadores relacionales para determinar si se cumple o no una relación de orden, por lo que el resultado de una comparación devuelve un valor de tipo `bool`.

````python
# Comparacion
a, b = 5, 7.3
print(a, '==', b, 'es:', a == b)
print(a, '!=', b, 'es:', a != b)
print(a, '>', b,'es:', a > b)
print(a, '>=', b, 'es:', a >= b)
print(a, '<', b, 'es:', a < b)
print(a, '<=', b, 'es:', a <= b)
````

Para el caso de cadenas (`str`), es posible comparar si dos cadenas son iguales mediante el operador `==` o diferentes con `!=`. Pero incluso podemos hacer comparaciones entre cadenas ya que el orden está establecido a través del orden alfabético.

````python
# Ejemplo del uso de la comparacion entre cadenas
palabra1 = 'Anaconda'
palabra2 = 'Raton'

print(palabra1 < palabra2)
print(palabra1 > palabra2)
print(palabra1 == palabra2)
````

La manera que se hace esto es a través de los valores en la codificación de los caracteres en la computadora. Python utiliza preferentemente la codificación UTF-8. Así, se van comparando carácter por carácter a través de sus valores enteros y de esta manera se determina cual está primero alfabéticamente.

**Nota**: Podemos usar la función `ord` para recuperar el valor entero de un carácter. Y podemos hacer lo contrario usando la función `chr` para una entrada entera.

````python
# Uso de la funcion ord
print(ord('A'))
print(ord('a'))
print(ord('Á'))
print(ord('á'))
````

````python
# Uso de la funcion chr
print(chr(ord('A') + 0))
print(chr(ord('A') + 1))
print(chr(ord('A') + 2))
print(chr(ord('A') + 3))
````

````python
# Tambien es posible utilizar su valor hexadecimal dada por la codificacion
print(chr(0x706B))
````

Sin embargo, se pueden utilizar los *literales de cadena formateados* (*formatted string literal*) o ***f-string***, para facilitar la lectura desde el código. Los *f-string* son un literal de cadena que se prefija con `f' '` o `F' '`, y son cadenas que pueden contener campos de reemplazo, las cuales son expresiones delimitadas por llaves `{}`. El siguiente ejemplo muestra su uso.

````python
a, b, c = 29, 7, 18
print('Los números ganadores son:', a, ',', b, 'y', c)
print(f'Los números ganadores son: {a}, {b} y {c}')
````

También se pueden insertar expresiones de Python entre las llaves. Las expresiones se reemplazan con su resultado en la cadena, pero se recomienda que las expresiones sean simples. Por ejemplo,

````python
a, b = 8.5, 9.73
print(f'{a} x {b} = {a*b}')
````

Las *f-strings* solo están disponibles de la versión de Python 3.6 en adelante. En las versiones previas y actuales de Python, el método `.format()` de las cadenas se puede usar para obtener los mismos resultados.

````python
a, b = -15.2, 79.43
print('{} x {} = {}'.format(a, b, a*b))
````

En este caso, cada uno de las llaves vacías corresponde a cada una de las variables o expresiones en la parte del método `.format()`. En ambos casos es posible añadir un especificador de formato utilizando `:` para controlar como se muestran los valores de las variables en la cadena final.

````python
pi = 3.14159265359
print(f'Diferentes aproximaciones de pi son: {pi: 3.2f}, {pi: <15.4f}, o {pi: 2.11F}')
print('Diferentes aproximaciones de pi son: {0: 3.2f}, {0: =+15.4f}, o {0: 2.11f}'.format(pi))
print(f'{1000000000.00: ,.2f}')
````

En el caso de las *f-strings* es necesario poner la variable y seguido de los dos puntos el especificador de formato que se le aplicará a dicha variable. Si se utiliza el método `.format()` es necesario especificar la posición dentro de este (observe que aquí el primero está indicado por 0) y nuevamente después de los dos puntos el especificador de formato. La sintaxis general para el especificador de formato es:
```python
{:[bandera][ancho][.precision]tipo}
```

## Función input

Cuando se realiza un *programa* es necesario recibir información o valores para declarar las variables, es decir, asignarles un valor. Por lo general, esto se hace a través de pedir esta información al usuario. La función `input` permite desplegar mensajes y recibir información con la siguiente sintaxis:


```python
var_salida = input('Mensaje como cadena de texto')
```

Una cosa que hay que tener en cuenta es que el argumento de salida (`var_salida`) de `input` es una cadena (tipo `str`), por lo que dependiendo del uso hay que hacer una conversión (*casting*) en el tipo de dato. Esto se hace a través de las funciones

* `int(x)` Convierte el argumento `x` a uno de tipo `int`.
* `float(x)` Convierte el argumento `x` a uno de tipo `float`.
* `str(x)` Convierte el argumento `x` a uno de tipo `str`.

En el siguiente ejemplo se pedirá un valor al usuario para calcular su cuadrado.

````python
# Programa para calcular el cuadrado de un numero
n = input('Dame un numero para calcular su cuadrado: ')
print(type(n))
print(n + '^2 = ' + str(float(n)**2))
````

Independientemente del valor dado por el usuario `n` será un tipo `str`, como muestra la instrucción `print(type(n))`. Para poder calcular el cuadrado del valor en `n` es necesario convertir la cadena a un tipo numérico. En el programa se usa `float(n)` para convertir el valor de `n` a un tipo `float`, calcular su cuadrado con `float(n)**2`, y el resultado convertirlo nuevamente a una cadena con `str(float(n)**2))`, para mandar el mensaje usando concatenación de cadenas.

Por otro lado, la función `eval` es útil cuando se requiere ingresar valores que son el resultado de evaluar una expresión en lugar de usar literales. Para ello, la función `eval` analiza la expresión (dada como una cadena) pasada a esta y ejecuta la expresión (código) de Python dentro del programa. De esta manera, usada junto con `input`, podemos dar cualquier expresión valida de Python para declarar variables de los diferentes tipos de datos sin la necesidad de usar las otras funciones para conversión de tipos.

````python
t = eval(input('Dame una entrada: '))
print(t)
print(type(t))
````

## Funciones matemáticas

Para tener acceso a las funciones matemáticas que comúnmente se utilizan, y que se pueden encontrar por ejemplo en una calculadora científica, es necesario importar un módulo. Python es un lenguaje de propósito general, por lo que varias funciones para propósitos específicos vienen empaquetadas en módulos.

Una biblioteca o módulo de funciones adicionales se importa mediante el comando `import`. Este comando solo es necesario ejecutarlo una vez antes de usar cualquier función que esté en dicho módulo.


### Módulo `math`

En el módulo `math` vienen definidas las funciones matemáticas usuales como: funciones trigonométricas, raíz cuadrada, exponencial, logarítmos, redondeos, etc, además de constantes como $\pi$ o $e$. Para usar dichas funciones o constantes se usa la instrucción `import math` y después hay que utilizarlas a través del espacio de nombres `math`. Por ejemplo, para usar la función *seno*, es necesario usar el comando `math.sin()`. Esto indica que se usará la función seno definida en el módulo `math`. Se hace de este modo debido a que una función puede estar definida en múltiples módulos (como se verá más adelante) y es necesario indicar a qué función de qué módulo se hace referencia y evitar ambigüedades. La documentación del módulo `math` se puede consultar en la siguiente [liga](https://docs.python.org/3/library/math.html).

````python
# Importando el modulo math
import math
````

````python
# Usando la función coseno del modulo math
math.cos(0)
````

````python
# Mostrando el valor de la constante pi del modulo math
math.pi
````

````python
# Usando la función raiz cuadrada del modulo math
math.sqrt(2)
````

````python
# Usando la función techo del modulo math
math.ceil(3.45)
````

## Ejemplos

### Ejemplo 1

1. Evalúe la siguiente expresión y muestre el resultado.
$$
    (-3.5)^{3} + \dfrac{e^6}{\ln 524} + 206^{1/3}
$$

**Solución**: Para calcular lo que se pide, se utiliza la función exponencial `math.exp` para evaluar el término $e^6$ y la función `math.log` para calcular el logaritmo natural.

````python
# Es necesario cargar el modulo math para cargar las funciones.
# Descomentar si no se ha cargado previamente.
# import math

expr = (-3.5)**3 + math.exp(6)/math.log(524) + 206**(1/3)
print(f'El resultado es: {expr: 2.6f}')
````

2. Evalúe la siguiente expresión y muestre el resultado.
$$
    \frac{2.5^{3}\left(16-\dfrac{216}{22}\right)}{1.7^{4}+14}+\sqrt[4]{2050}.
$$

**Solución**: Es importante tener en cuenta que $\sqrt[4]{x} = x^{\frac{1}{4}}$. En este caso en lugar del operador `**`, se uso la función `math.pow` para calcular este término.

````python
# Es necesario cargar el modulo math para cargar las funciones.
# Descomentar si no se ha cargado previamente.
# import math

expr = 2.5**3 * (16 - 216/22) / (1.7**4 + 14) + math.pow(2050, 1/4)
print(f'El resultado es: {expr: 2.6f}')
````

3. Evalúe la siguiente expresión y muestre el resultado.
$$
    \dfrac{\tan 64^{\circ}}{\cos^{2}14^{\circ}} - \dfrac{3\sin 80^{\circ}}{\sqrt[3]{0.9}} + \dfrac{\cos 55^{\circ}}{\sin 11^{\circ}}.
$$

**Solución**: Como la expresión contempla los ángulos medidos en grados, es necesario convertirlos a radianes, ya que las funciones trigonométricas dentro del módulo `math` reciben los argumentos en radianes. Para ello se puede usar la función `math.radians` para hacer la conversión de grados a radianes. También existe la función `math.degrees` para el proceso inverso. Dado que la expresión es muy larga, para partirla en múltiples líneas se puede usar el símbolo `\` al final de cada línea. Esto le índica a Python que la instrucción continua en la siguiente línea, y de este modo la instrucción es legible.

````python
# Es necesario cargar el modulo math para cargar las funciones.
# Descomentar si no se ha cargado previamente.
# import math

expr = math.tan(math.radians(64)) / math.cos(math.radians(14))**2 \
       - 3*math.sin(math.radians(80)) / math.pow(0.9, 1/3) \
       + math.cos(math.radians(55)) / math.sin(math.radians(11))
print(f'El resultado es: {expr: 2.6f}')
````

4. Defina las variables $a$, $b$, $c$ y $d$ como: $a = 12$, $b = 5.6$, $c = \dfrac{3a}{b^2}$, y $d = \dfrac{(a-b)^{c}}{c}$, evalúe la siguiente expresión y muestre el resultado
$$
    \exp\left(\dfrac{d-c}{a-2b}\right) + \ln\left(\left|c-d+\dfrac{b}{a}\right|\right)
$$

**Solución**: Aquí solo hay tener cuidado en la asignación de los valores a las variables y usar la función ```math.fabs()``` para calcular el valor absoluto.

````python
# Fase de inicializacion
# Es necesario cargar el modulo math para cargar las funciones.
# Descomentar si no se ha cargado previamente.
# import math
a, b = 12, 5.6

# Fase de procesamiento
c = 3*a / b**2
d = (a-b)**c / c

# Fase de terminacion
expr = math.exp((d-c) / (a - 2*b)) + math.log(math.fabs(c - d + b/a))
print(f'El resultado es: {expr: 2.6f}')
````
