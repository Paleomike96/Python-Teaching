### Ejercicio 1

Escribe las matrices faltantes como un arreglo en Python y eleva todas las matrices a la potencia n con n = 10, 100, 1000.

Las matrices son:

\[
P1 = \begin{pmatrix}
1/2 & 1/2 & 0 \\
1/2 & 0 & 1/2 \\
0 & 1/2 & 1/2  \\
\end{pmatrix}
\]

\[
P2 = \begin{pmatrix}
1/2 & 1/2 & 0 \\
1/2 & 99/200 & 1/200 \\
0 & 0 & 1  \\
\end{pmatrix}
\]

\[
P3 = \begin{pmatrix}
0 & 1 & 0 & 0 & 0 & 0 \\
1/5 & 0 & 4/5 & 0 & 0 & 0 \\
0 & 2/5 & 0 & 3/5 & 0 & 0 \\
0 & 0 & 3/5 & 0 & 2/5 & 0 \\
5 & 0 & 0 & 4/5 & 0 & 1/5 \\
0 & 0 & 0 & 0 & 1 & 0 \\
\end{pmatrix}
\]

\[
P4 = \begin{pmatrix}
1 & 0 & 0 & 0 & 0 & 0 \\
0.5 & 0 & 0.5 & 0 & 0 & 0 \\
0 & 0.5 & 0 & 0.5 & 0 & 0 \\
0 & 0 & 0.5 & 0 & 0.5 & 0 \\
0 & 0 & 0 & 0.5 & 0 & 0.5 \\
0 & 0 & 0 & 0 & 0 & 1 \\
\end{pmatrix}
\]


````python
#Importamos las librerias
import numpy as np

# Definimos las matrices P1, P2, P3, P4
P1 = np.array([[1/2, 1/2, 0],
               [1/2, 0, 1/2],
               [0, 1/2, 1/2]])

P2 = np.array([[1/2, 1/2, 0],
               [1/2, 99/200, 1/200],
               [0, 0, 1]])

P3 = np.array([[0, 1, 0, 0, 0, 0],
               [1/5, 0, 4/5, 0, 0, 0],
               [0, 2/5, 0, 3/5, 0, 0],
               [0, 0, 3/5, 0, 2/5, 0],
               [5, 0, 0, 4/5, 0, 1/5],
               [0, 0, 0, 0, 1, 0]])

P4 = np.array([[1, 0, 0, 0, 0, 0],
               [0.5, 0, 0.5, 0, 0, 0],
               [0, 0.5, 0, 0.5, 0, 0],
               [0, 0, 0.5, 0, 0.5, 0],
               [0, 0, 0, 0.5, 0, 0.5],
               [0, 0, 0, 0, 0, 1]])

# Definimos el valor de n
n_values = [10, 100, 1000]

# Función para elevar una matriz a diferentes potencias
def matrix_powers(matrix, n_values):
    results = {}
    for n in n_values:
        results[n] = np.linalg.matrix_power(matrix, n)
    return results

# Elevamos todas las matrices a las potencias 10, 100 y 1000
P1_powers = matrix_powers(P1, n_values)
P2_powers = matrix_powers(P2, n_values)
P3_powers = matrix_powers(P3, n_values)
P4_powers = matrix_powers(P4, n_values)

# Mostramos los resultados para cada potencia de las matrices
for n in n_values:
    print(f"P1 elevado a la potencia {n}:\n", P1_powers[n], "\n")
    print(f"P2 elevado a la potencia {n}:\n", P2_powers[n], "\n")
    print(f"P3 elevado a la potencia {n}:\n", P3_powers[n], "\n")
    print(f"P4 elevado a la potencia {n}:\n", P4_powers[n], "\n")
````

### Ejercicio 2

Haz una función que simule un estado dado una distribución de una Cadena de Markov considerando una distribución inicial.

Dado un vector de probabilidades $ v = (x_1, x_2, \dots, x_n) $, tu función debe recibir:
- Un vector de probabilidades cuya suma sea 1.

Debe regresar de manera aleatoria (con la probabilidad de elección correspondiente) una de las etiquetas de los estados.


````python
import numpy as np

# Función que simula un estado de una cadena de Markov
def simular_estado(v, etiquetas):
    """
    Simula un estado dado un vector de probabilidades y etiquetas de estados.

    Parameters:
    v: lista o array de probabilidades cuya suma sea 1
    etiquetas: lista de etiquetas correspondientes a los estados

    Returns:
    Una de las etiquetas de los estados, seleccionada de manera aleatoria
    según la distribución de probabilidades dada en v.
    """
    # Verificamos que la suma de probabilidades sea 1
    if not np.isclose(np.sum(v), 1):
        raise ValueError("La suma de las probabilidades debe ser 1.")

    # Simulamos el estado con las probabilidades correspondientes
    estado_simulado = np.random.choice(etiquetas, p=v)

    return estado_simulado

# Ejemplo de uso
v = [0.2, 0.5, 0.3]  # Vector de probabilidades
etiquetas = ['Estado 1', 'Estado 2', 'Estado 3']  # Etiquetas de los estados

# Simulamos el estado
estado = simular_estado(v, etiquetas)
print(f"Estado simulado: {estado}")
````

### Ejercicio 3

En este ejercicio programarán dos funciones que simulen una cadena de Markov y los estados que va tomando en cada iteración. La diferencia entre estas funciones será el criterio de paro.

- **Primera función**:
  1. Debe realizar N iteraciones de la cadena de Markov y recibir este valor como parámetro.
  2. Debe regresar la lista de los estados que recorrió.

- **Segunda función**:
  1. Debe correr hasta que la cadena tome un estado \( e \).
  2. Debe recibir la etiqueta del estado al que se quiere llegar y un número máximo de iteraciones, \( \text{MaxIter} \), en caso de que nunca se alcance el estado deseado.

Ambas funciones deben recibir dos parámetros correspondientes a la cadena:
- El primer parámetro será la matriz de transición de la cadena, \( P \).
- El segundo parámetro será la distribución inicial de la cadena, \( \Pi_0 \).


````python
import numpy as np

# Función 1: Simular N iteraciones de la cadena de Markov
def simular_cadena_markov_N_iteraciones(P, Pi_0, N):
    """
    Simula una cadena de Markov con N iteraciones.
    Parameters:
    P: Matriz de transición de la cadena de Markov
    Pi_0: Distribución inicial
    N: Número de iteraciones
    Returns:
    Una lista con los estados recorridos.
    """
    estados = []
    estado_actual = np.random.choice(len(Pi_0), p=Pi_0)  # Seleccionar el estado inicial
    estados.append(estado_actual)

    for _ in range(N):
        estado_actual = np.random.choice(len(P), p=P[estado_actual])
        estados.append(estado_actual)
    return estados

# Función 2: Simulamos hasta alcanzar el estado dado o cierto número máximo de iteraciones
def simular_cadena_markov_hasta_estado(P, Pi_0, e, MaxIter):
    """
    Simula una cadena de Markov hasta alcanzar un estado e o hasta que se cumplan MaxIter iteraciones.
    Parameters:
    P: Matriz de transición de la cadena de Markov
    Pi_0: Distribución inicial
    e: Estado objetivo
    MaxIter: Número máximo de iteraciones
    Returns:
    Una lista con los estados recorridos.
    """
    estados = []
    estado_actual = np.random.choice(len(Pi_0), p=Pi_0)  # Seleccionar el estado inicial
    estados.append(estado_actual)
    for _ in range(MaxIter):
        if estado_actual == e:
            break
        estado_actual = np.random.choice(len(P), p=P[estado_actual])
        estados.append(estado_actual)
    return estados

# Ejemplo de uso:
# Matriz de transición P
P = np.array([[0.5, 0.5, 0],
              [0.25, 0.5, 0.25],
              [0, 0.5, 0.5]])
# Distribución inicial Pi_0
Pi_0 = [0.2, 0.5, 0.3]
# Número de iteraciones
N = 10
# Estado objetivo
estado_objetivo = 2
MaxIter = 100
# Simulación de N iteraciones
estados_N = simular_cadena_markov_N_iteraciones(P, Pi_0, N)
print(f"Estados recorridos en {N} iteraciones: {estados_N}")
# Simulación hasta alcanzar un estado dado o MaxIter
estados_hasta_e = simular_cadena_markov_hasta_estado(P, Pi_0, estado_objetivo, MaxIter)
print(f"Estados recorridos hasta alcanzar el estado {estado_objetivo} o {MaxIter} iteraciones: {estados_hasta_e}")
````

Podemos aproximar la probabilidad de que la cadena se encuentre en cada estado después de muchas iteraciones usando la distribución inicial $\Pi_0$ y la matriz de transición $P$. Esta aproximación es llamada la **distribución límite**.

Para ello, basta hacer el producto $\Pi_0 P^n$ para valores de $n$ grandes. Haz ese producto para cada matriz de transición utilizando $n = 1000$ y tres distribuciones iniciales generadas de manera aleatoria.

**Pregunta**: ¿La elección de la distribución inicial afecta en algo?


````python
import numpy as np

# Función para calcular la distribución límite
def calcular_distribucion_limite(P, Pi_0, n):
    """
    Calcula la distribución límite de una cadena de Markov.
    Parameters:
    P: Matriz de transición de la cadena de Markov.
    Pi_0: Distribución inicial.
    n: Número de iteraciones.
    Returns:
    Distribución límite después de n iteraciones.
    """
    # Elevar la matriz de transición P a la potencia n
    P_n = np.linalg.matrix_power(P, n)
    # Calcular la distribución límite Pi_0 * P^n
    Pi_n = np.dot(Pi_0, P_n)
    return Pi_n
# Matrices de transición del Ejercicio 1
P1 = np.array([[0.5, 0.5, 0],
               [0.5, 0, 0.5],
               [0, 0.5, 0.5]])
P2 = np.array([[0.5, 0.5, 0],
               [0.5, 99/200, 1/200],
               [0, 0, 1]])
P3 = np.array([[0, 1, 0, 0, 0, 0],
               [1/5, 0, 4/5, 0, 0, 0],
               [0, 2/5, 0, 3/5, 0, 0],
               [0, 0, 3/5, 0, 2/5, 0],
               [5/5, 0, 0, 4/5, 0, 1/5],
               [0, 0, 0, 0, 1, 0]])
P4 = np.array([[1, 0, 0, 0, 0, 0],
               [0.5, 0, 0.5, 0, 0, 0],
               [0, 0.5, 0, 0.5, 0, 0],
               [0, 0, 0.5, 0, 0.5, 0],
               [0, 0, 0, 0.5, 0, 0.5],
               [0, 0, 0, 0, 0, 1]])
# Generar distribución inicial aleatoria
def generar_distribucion_inicial(n):
    Pi_0 = np.random.random(n)  # Generar valores aleatorios entre 0 y 1
    Pi_0 /= Pi_0.sum()  # Convertirlos en una distribución de probabilidad
    return Pi_0
# Número de iteraciones grandes (n = 1000)
n_iter = 1000
# Generar tres distribuciones iniciales aleatorias
Pi_0_1 = generar_distribucion_inicial(P1.shape[0])
Pi_0_2 = generar_distribucion_inicial(P2.shape[0])
Pi_0_3 = generar_distribucion_inicial(P3.shape[0])
Pi_0_4 = generar_distribucion_inicial(P4.shape[0])
# Calcular la distribución límite para cada matriz con la primera distribución inicial
distribucion_limite_P1 = calcular_distribucion_limite(P1, Pi_0_1, n_iter)
distribucion_limite_P2 = calcular_distribucion_limite(P2, Pi_0_2, n_iter)
distribucion_limite_P3 = calcular_distribucion_limite(P3, Pi_0_3, n_iter)
distribucion_limite_P4 = calcular_distribucion_limite(P4, Pi_0_4, n_iter)
# Imprimir las distribuciones límite
print("Distribución límite para P1:", distribucion_limite_P1)
print("Distribución límite para P2:", distribucion_limite_P2)
print("Distribución límite para P3:", distribucion_limite_P3)
print("Distribución límite para P4:", distribucion_limite_P4)
# Compara si la elección de la distribución inicial afecta (pistas para el análisis final)
# Podemos observar si las distribuciones límites son similares independientemente de la distribución inicial
````
