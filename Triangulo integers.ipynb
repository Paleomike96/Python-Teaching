```python
def crear_piramide_invertida(n, unicode_codigo):
    #Convertimos del codigo UNINCODE al caracter correspondiente
    caracter = chr(unicode_codigo)

    #Creamos o más bien, imprimimos nuestra piramide invertida
    for i in range(n, 0, -1):
        #Agregamos los espacios para centrar nuestro caracter
        espacios = ' ' * (n - i)
        #Imprimimos la línea de nuestra piramide
        print(espacios + ' '.join([caracter] * i))

#En esta parte comienza nuestro programita solicitado
def main():
    while True:
#Solicitamos el número de niveles (un número) y el código UNICODE (número) al usuario
        n_str = input("Ingrese el número de niveles (n): ")
        unicode_codigo_str = input("Ingrese el código UNICODE del carácter: ")

        #Validamos en el programa nuestro número de niveles y el codigo UNICODE, de forma que sean enteros
        if n_str.isdigit() and unicode_codigo_str.isdigit():
            n = int(n_str)
            unicode_codigo = int(unicode_codigo_str)

            #Aseguramos que no escriba el usuario niveles 0 o negativos
            if n < 1:
                print("El número de niveles debe ser al menos 1.")
                continue

            #Creamos nuestra piramide invertida
            crear_piramide_invertida(n, unicode_codigo)

            #Preguntamos al usuario si desea seguir en el programa o salir de este
            respuesta = input("¿Desea ingresar nuevos parámetros? (s/n): ").strip().lower()
            if respuesta != 's':
                print("Terminando el programa.")
                break
        else:
            print("Por favor, ingrese valores numéricos válidos.")

#Termina nuestro programa
if __name__ == "__main__":
    main()
```
