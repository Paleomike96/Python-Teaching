```python
import numpy as np
import matplotlib.pyplot as plt

# CODIGO PARA GRAFICAR FUNCIONES EN COORDENADAS PORLARES #
theta = np.linspace(0,2*np.pi)
r = 5 + 50*theta

fig = plt.figure()
ax = fig.add_subplot(111, projection="polar",facecolor="skyblue")
ax.plot(theta,r)

plt.show()
```

```python
#CODIGO PARA PASAR FUNCIONES DE COORDENADAS POLARES A CARTESIANAS
import math
def polar_cart(r,theta):
  x = r*round(math.cos(math.radians(theta)),2)
  y = r*round(math.sin(math.radians(theta)),2)
  print(f"El valor de x es:{x} y de y es: {y}")

#LLAMADA DE LA FUNCIOM #
polar_cart(1,90)
```

```python
#CODGIO QUE GRAFICA FUNCIONES EN COORDENADAS CARTESIANAS
t = np.linspace(-10,10)
plt.plot(t,t**2,color="r")
plt.title("Gráfica de $x^2$")
plt.show()
```

```python
#Codigo que transforma coordenadas cartesianas a polares #
from numpy import arctan
def cart_pol (x,y):
  r = round(math.sqrt(x**2+y**2),2)
  if x != 0:
    theta = round((arctan(y/x))*180/np.pi,2)
  elif x == 0 and y>0:
    theta =  90
  elif x == 0 and y<0:
    theta = 270
  elif x==0 and y==0:
    theta = 0

  print(f"El valor de r es:{r} y de theta es: {theta}")
  plt.plot(x,y,color="r",marker="o")
  plt.title("Gráfica del punto")
  plt.show()

cart_pol(1,1)

```
