# Estrategia de Ahorro para la Compra de una Notepad

## Presupuesto para los artículos:
Dado que quiero comprarme una Notepad para mis clases de Mates aplicadas, me dedique a investigar precios en Amazon (ya que me resultan a mi criterio los que tienen mejores electrónicos, al contrario de Mercado Libre)

1. **Notepad Básica (Modelo económico)**: MXN $20,000

2. **Notepad de Gama Media (Mejores características)**: MXN $30,000

3. **Notepad de Gama Alta (Alta capacidad y rendimiento)**: MXN $50,000

 (Los precios son aproximados, ya que me gusta manejar datos discretos)

## Rango de ahorro e inversión inicial
- **Tiempo de ahorro**: 8 meses (en mi mundo ideal)

- **Ahorro mensual**: $8,000 MXN (considerando que me consigo buen trabajo de 'medio tiempo')

- **Inversión disponible**: Entre 80% y 100% del total de los ahorros. (Realmente solo gasto en servicios de Streaming)

- **Inversión mínima (80%)**: MXN $6,400 mensuales.

- **Inversión máxima (100%)**: MXN $8,000 mensuales.

## Escenarios de inversión
Ahora, construyo tres escenarios variando el monto de la **inversión inicial** y el **depósito mensual**:

1. **Escenario 1**: Inversión inicial de $5,000$ y depósito mensual de $6,400.

2. **Escenario 2**: Inversión inicial de $10,000$ y depósito mensual de $7,200.

3. **Escenario 3**: Inversión inicial de $20,000$ y depósito mensual de $8,000.

## Análisis de los Escenarios

### Escenario 1:
- **Inversión inicial**: $5,000 MXN.

- **Depósitos mensuales**: $6,400 MXN.

- **Resultado**: El saldo acumulado es suficiente para la Notepad de gama baja (~$20,000 MXN), pero no me alcanza para las gamas media o alta.

### Escenario 2:
- **Inversión inicial**: $10,000 MXN.

- **Depósitos mensuales**: $7,200 MXN.

- **Resultado**: El saldo acumulado es suficiente para la Notepad de gama media (~$30,000 MXN), pero no es suficiente para la de gama alta (aún :'( ).

### Escenario 3:
- **Inversión inicial**: $20,000 MXN.

- **Depósitos mensuales**: $8,000 MXN.

- **Resultado**: El saldo acumulado es suficiente para la Notepad de gama alta (~$50,000 MXN), logrando mi objetivo más ambicioso.


```python
#Ahora implementamos el código de sistemas dinamicos de la forma P_{n+1} = aP_n + b para mi supuesto de la Notepad
#Importamos las librerias a utilizar
import numpy as np
import matplotlib.pyplot as plt

# Definimos el modelo de interés compuesto con depósitos mensuales
def ahorro_cetes(inversion_inicial, deposito_mensual, tasa_anual, meses):
    tasa_mensual = tasa_anual / 12 / 100  # Convertimos tasa anual a mensual
    ahorro = np.zeros(meses + 1)
    ahorro[0] = inversion_inicial
    for n in range(meses):
        ahorro[n+1] = ahorro[n] + deposito_mensual + tasa_mensual * ahorro[n]
    return ahorro #Mostramos nuestro ahorro dadas las condiciones de intereses de CETES y mis supuestos de ahorro

# Parámetros a utilizar
tasa_anual = 11  # Tasa anual en 2024
meses = 8  #Mi tiempo de ahorro

#Escenarios de inversión de acuerdo a mis capacidades
escenario_1 = ahorro_cetes(5000, 6400, tasa_anual, meses)  # Escenario 1
escenario_2 = ahorro_cetes(10000, 7200, tasa_anual, meses)  # Escenario 2
escenario_3 = ahorro_cetes(20000, 8000, tasa_anual, meses)  # Escenario 3

# Graficamos los resultados
plt.figure(figsize=(10, 6))
plt.plot(escenario_1, label='Escenario 1: $5,000 inicial, $6,400 mensual')
plt.plot(escenario_2, label='Escenario 2: $10,000 inicial, $7,200 mensual')
plt.plot(escenario_3, label='Escenario 3: $20,000 inicial, $8,000 mensual')
plt.title('Evolución del ahorro con CETESDIRECTO en 8 meses')
plt.xlabel('Meses')
plt.ylabel('Saldo acumulado (pesos)')
plt.legend()
plt.grid(True)
plt.show()

# Comparación final de los distintos escenarios, y poder visualizar graficamente cual estrategia es la más óptima
print(f"Saldo acumulado en Escenario 1: ${escenario_1[-1]:.2f}")
print(f"Saldo acumulado en Escenario 2: ${escenario_2[-1]:.2f}")
print(f"Saldo acumulado en Escenario 3: ${escenario_3[-1]:.2f}")
```

## Estrategia recomendada
Para comprar la **Notepad de gama alta** en 8 meses, el **Escenario 3** con una inversión inicial de $20,000 MXN$ y depósitos mensuales de $8,000 resulta ser el mejor, o al menos el más atractivo, suponiendo mi inversion de intereses en CETES (aunque analizandolo bien, esta práctica también podria aplicar a NU banco)

Aunque la **Notepad de gama media** del **Escenario 2** me permite ahorrar más que suficiente, y no tener que dar una inversión demasiado grande
